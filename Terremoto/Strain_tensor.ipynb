### Cómo se pueden interpretar estos datos?

EN el campo de la geología estructural hablamos de deformación y esfuerzo. SOn dos factores que van relacionados pero SON DIFERENTES.
La deformación es el producto final que podemos VER y MEDIR de un evento tectónico que se registra en las rocas. MIentras que los esfuerzos son  instantaneos y no los podemos ver ni medir directamente en una roca.
Por eso, al poder medir la deformación, estamos caracterizando esta misma. COn esto podemos saber cual fue el eje de mayor deformación.

la deformación , asi como el esfuerzo, tiene 3 ejes principales (d1, d2, y d3) en el cual d1 es la dirección de mayor deformación y d3 el de la menor.

con estos resultados, al tener el autovector y el autovalor principal (que es el mas grande, en el cual el autovalor es el mayor)
podemos saber la dirección y magnitud de la deformación principal que sufrio al tierra. EN este caso Chile después del terremoto
