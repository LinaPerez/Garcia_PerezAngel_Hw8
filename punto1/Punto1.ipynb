```python
filename = 'ves3.tif'
imagen = Image.open(filename) #http://effbot.org/imagingbook/image.htm
    
array_imagen = numpy.asarray(imagen) # Convert the input to an array. - http://docs.scipy.org/doc/numpy/reference/generated/numpy.asarray.html

imshow(array_imagen)

shape(array_imagen)
```

```python
array_imagen=(1.0*array_imagen)/(amax(array_imagen)-amin(array_imagen))#Normalizacion de la imagen
```

```python
def par(r0,r1,m,phi,theta):
    return r0+r1*sin(m*(theta)+phi) #Se define la parametrizacion propuesta
```

```python
theta1=linspace(0,2*pi,1000)
phi1=linspace(0,2*pi,1000)
def my_model(r0,r1,m,phi):
    a=par(r0,r1,m,phi,theta1)*cos(theta1)
    b=par(r0,r1,m,phi,phi1)*sin(phi1)
    return [a,b]
```

El proximo algoritmo convierte la parametrizacion a una imagen de la misma dimension que la original

```python
px=shape(array_imagen)[0]
py=shape(array_imagen)[1]
def run(a,b,x0,y0):
    nx=int(shape(array_imagen)[0])
    ny=int(shape(array_imagen)[1])
    figure = numpy.zeros((nx, ny))
    for i in range(size(b)):
        figure[(round(a[i])+round(y0))%px][(round(b[i])+round(x0))%py]=1
    return figure
        
```

```python
y_obs=array_imagen
```

Definimos una funcion de likelihood. Despues de probar muchisimas con exponencial la mejor que se comporta es esta. El problema de usar exp(-chi_squared) es que se debe conocer de alguna forma el error o dispercion de los datos.

```python
def likelihood(y_obs, y_model):
    chi_squared = (1.0/2.0)*sum((y_obs-y_model)**2)
    return (-chi_squared)
```

En la proxima parte del codigo inicializamos de donde va a arrancar el algoritmo MCMC. Proponemos algunos limites para cada parametro

```python
#Corresponde a los pasos del algoritmo, Si el radio es menor que a un valor le damos probabilidad 0
n_iterations = 10000 #this is the number of iterations I want to make
for i in range(n_iterations):
    
    r0_prime = np.random.normal(r0_walk[i], 0.1)   
    r1_prime = np.random.normal(r1_walk[i], 0.1)
    m_prime = np.random.normal(m_walk[i], 0.1)
    phi_prime = np.random.normal(phi_walk[i], 0.1)
    x0_prime = np.random.normal(x0_walk[i], 0.7)
    y0_prime = np.random.normal(y0_walk[i], 0.7)
    
    y_init1 = my_model(r0_walk[i], r1_walk[i], m_walk[i], phi_walk[i])
    y_init = run(y_init1[0],y_init1 [1],x0_walk[i],y0_walk[i] )
    
    y_prime1 = my_model(r0_prime, r1_prime, m_prime, phi_prime)
    y_prime = run(y_prime1[0],y_prime1 [1], x0_prime, y0_prime )
    
    l_prime = likelihood(y_obs, y_prime)
    l_init = likelihood(y_obs, y_init)
    
    
    alpha = l_prime/l_init
    
    if(alpha<=1.0  ):
        r0_walk  = append(r0_walk,r0_prime)
        r1_walk  = append(r1_walk,r1_prime)
        m_walk = append(m_walk, m_prime)
        phi_walk = append(phi_walk, phi_prime)
        l_walk = append(l_walk, l_prime)
        x0_walk  = append(x0_walk,x0_prime)
        y0_walk  = append(y0_walk,y0_prime)
    else:
        beta = random.random()
        if(beta>=exp(-alpha) ):
            r0_walk  = append(r0_walk,r0_prime)
            r1_walk  = append(r1_walk,r1_prime)
            m_walk = append(m_walk, m_prime)
            phi_walk = append(phi_walk, phi_prime)
            l_walk = append(l_walk, l_prime)
            x0_walk  = append(x0_walk,x0_prime)
            y0_walk  = append(y0_walk,y0_prime)
        else:
            r0_walk = append(r0_walk,r0_walk[i])
            r1_walk = append(r1_walk,r1_walk[i])
            m_walk = append(m_walk, m_walk[i])
            phi_walk = append(phi_walk,phi_walk[i])
            l_walk = append(l_walk, l_init)
            x0_walk = append(x0_walk,x0_walk[i])
            y0_walk = append(y0_walk,y0_walk[i])
```
